```python
df.isnull().sum()
```

```python
# Fill missing numerical values
df['discount'] = df['discount'].fillna(df['discount'].mean())
# Fill missing text values
df['seller_name'] = df['seller_name'].fillna('Unknown')
df['seller_information'] = df['seller_information'].fillna('Not Available')
df['what_customers_said'] = df['what_customers_said'].fillna('No Review')
df['variations'] = df['variations'].fillna('Not Available')
df['videos'] = df['videos'].fillna('No Video')
```

```python
df.isnull().sum()
```

```python
df= df.dropna()   # rows with any missing values get dropped #nothing dropped as missing values have been filled
```

```python
df.isnull().sum()
```

```python
selected_columns = df[['title','category','rating']]
print(selected_columns.head())
```

```python
#Filter rows where rating is less than 3
low_rating_products = df[df['rating']<3]
low_rating_products.head()
```

```python
#Filter rows where category is Earrings
fashion_prods = df[df['category'] == 'earrings']
fashion_prods.head()
```

```python
initial_rows= df.shape[0]

df= df.drop_duplicates()
final_rows= df.shape[0]

print(f"Rows before:{initial_rows}")
print(f"Rows after removing duplicates:{final_rows}")
```

```python
df['total_amount']=  df['rating'] *df['ratings_count']
df[['rating','ratings_count','total_amount']].head()
```

```python
cleaned_file = 'cleaned_shopping_dataset.csv'
df.to_csv(cleaned_file, index=False)

print(f"Cleaned dataset saved as: {cleaned_file}")
```
